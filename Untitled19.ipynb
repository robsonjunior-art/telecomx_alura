```python
import pandas as pd
```

```python
df = pd.read_json('/content/TelecomX_Data.json')
print(df)
```

```python
df.info()
```

```python
customer_df = pd.json_normalize(df['customer'])
phone_df = pd.json_normalize(df['phone'])
internet_df = pd.json_normalize(df['internet'])
account_df = pd.json_normalize(df['account'])
df_final = pd.concat(
    [
        df[['customerID', 'Churn']],
        customer_df,
        phone_df,
        internet_df,
        account_df
    ],
    axis=1
)

print(df_final.head())
```

```python
df_final.info()
df_final.head()
```

```python
df_final['Charges.Total'] = pd.to_numeric(df_final['Charges.Total'], errors='coerce')
```

```python
col_cat = df_final.select_dtypes(include='object').columns.tolist()

col_cat = [col for col in col_cat if col != 'customerID']

col_cat
```

```python
def churn_by_category(df, cat_col):
    plt.figure(figsize=(8,4))
    sns.countplot(data=df, x=cat_col, hue='Churn', palette=['#2ca02c','#d62728'])
    plt.title(f'Churn por {cat_col}')
    plt.xlabel(cat_col)
    plt.ylabel('Número de Clientes')
    plt.legend(title='Churn', labels=['Não (0)', 'Sim (1)'])
    plt.xticks(rotation=45)
    plt.show()
```

```python
churn_by_category(df_final, 'gender')
churn_by_category(df_final, 'Contract')
churn_by_category(df_final, 'PaymentMethod')
churn_by_category(df_final, 'InternetService')
```

```python
for col in col_cat:
    prop = pd.crosstab(df_final[col], df_final['Churn'], normalize='index') * 100
    print(f'\nProporção de Churn por {col} (%):\n', prop.round(2))
```

O objetivo desta análise é entender a evasão de clientes de uma empresa de telecomunicações e identificar padrões relacionados à evasão (Churn).

### 3.2 Variáveis Numéricas
Boxplots e distribuições das variáveis de consumo por Churn:

```python
#Grafico 3.2
num_cols = ['Charges.Total', 'Charges.Monthly', 'tenure', 'Contas_Diarias']

plt.figure(figsize=(12,8))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=df_final, x='Churn', y=col, palette=['#2ca02c','#d62728'])
    plt.title(f'{col} x Churn')
    plt.xlabel('Churn')
    plt.ylabel(col)
plt.tight_layout()
plt.show()
```

4. Conclusões e Insights

Contratos mensais e clientes com menos tempo de permanência são mais propensos a cancelar.

Gasto baixo ou irregular (diário ou mensal) está associado à maior evasão.

Métodos de pagamento e tipo de serviço contratado também influenciam na retenção.

Há outliers que representam clientes de alto gasto que também cancelaram, indicando possíveis oportunidades de retenção.

5. Recomendações

Focar em retenção de clientes de contrato mensal: oferecer promoções ou fidelização.

Monitorar gastos e padrões de consumo: identificar clientes de baixo gasto antes que cancelem.

Aprimorar serviços adicionais (internet, telefone, suporte) para reduzir insatisfação.

Campanhas personalizadas por perfil de cliente: gênero, dependentes, serviços contratados.

Criar alertas para clientes em risco baseado em tenure e consumo.
